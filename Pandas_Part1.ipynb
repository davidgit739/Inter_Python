```python
dict = {
    "country":["Brazil", "Russia", "India", "China", "South Africa"],
    "capital":["Brasilia", "Moscos", "New Delhi", "Beijing", "Pretoria"], 
    "area": [8.516, 17.10, 3.286, 9.597, 1.221],
    "population":[200.4, 143.5, 1252, 1357, 52.98]}
```

```python
import pandas as pd 
brics = pd.DataFrame(dict)
```

```python
#to remove the first column index set index_col = 0 
#when importing data

#pd.read_csv("path/to/brics.csv", index_col = 0)
#The DataFrame is one of Pandas' most important data structures. 
#It's basically a way to store tabular data where you can label the rows and the columns. 
#One way to build a DataFrame is from a dictionary.

```

```python
# Pre-defined lists
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]

# Import pandas as pd
import pandas as pd 

# Create dictionary my_dict with three key:value pairs: my_dict
my_dict = {
    'country':names,
    'drives_right':dr,
    'cars_per_cap':cpc
}
#square brackets not needed above, this caused the lists to be stored as strings almost. Not seprate entities within the table. 
# Build a DataFrame cars from my_dict: cars
cars = pd.DataFrame(my_dict) 


# Print cars
print(cars)
```

```python
# Definition of row_labels
row_labels = ['US', 'AUS', 'JPN', 'IN', 'RU', 'MOR', 'EG']

# Specify row labels of cars
cars.index = row_labels

# Print cars again
print(cars) 
```

```python
#index and select data from data frames 
#loc label-based and 
#iloc integer position-based 
brics.loc["RU"]

```

```python
#above is a series. to get a data frame
brics.loc[["RU"]]
```

```python
brics.loc[["RU", "IN", "SA"]]
```

```python
#above grabs whole row. 
#row and column loc
brics.loc[["RU", "IN", "SA"],["country","capital"]] # by adding in the extra series of data labels we can specify what columns to return

```

```python
brics.loc[:,["country","capital"]] # all rows specified columns 
```
